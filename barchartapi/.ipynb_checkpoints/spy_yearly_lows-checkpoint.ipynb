Yearly lows in S&P 500
==================

### 1.01 Define imports and add barchartapi to sys.path

```python
import pandas as pd
import numpy as np
import os
import sys
import datetime
from os.path import expanduser
import matplotlib.pyplot as plt
%matplotlib inline

#add barchartapi project folder (..) and barchartapi package folder (../barchartapi) to sys.path
PROJECT_DIR = f'..' 
PACKAGE_DIR = f'{PROJECT_DIR}/barchartapi' 
if PROJECT_DIR not in sys.path:
    sys.path.append(PROJECT_DIR)
if PACKAGE_DIR not in sys.path:
    sys.path.append(PACKAGE_DIR)
import barchart_api as bcapi
import futures_series as futs
```

### 2.01  Create an instance of BcHist 

BcHist allows you to fetch daily history of SPY using the Barchart API

```python

api_key = open('./temp_folder/paid_api_key.txt','r').read()
bar_type='daily'
interval=1
endpoint_type='paid_url'
bch = bcapi.BcHist(api_key, bar_type=bar_type, interval=interval,endpoint_type = endpoint_type)
```

### 3.01 Fetch the data

```python
# define name of ETF
short_name = 'SPY'
# For begin and end dates, 
#   define the year as 4 digits, the month as 2 digits and the day as 2 digits
#   and concatenate year, month and day into one integer
beg_yyyymmdd = 19000101
end_yyyymmdd = 20181231
# call the get_history method.  
# get_history returns a tuple, where tup[0] is a status dictionary and
#    tup[1] is either a DataFrame with the returned data, or None
tup = bch.get_history(short_name, beg_yyyymmdd, end_yyyymmdd)
tup[0]
```

### 3.02  Put the data in the variable df

```python
df = tup[1]
df.head()
```

### 3.03 Execute proc_df (in module futures_series)

1. **proc_df** will add year, month and day columns.    
2. **proc_df** will also add other date oriented columns, *that we will drop*

```python
df_2 = futs.proc_df(df=df,is_commodity=False)
```

```python
df_2.head()
```

### 4.01 Create data frames that:

1. hold the first day in January for each year, and 
2. the last day in December for each year

```python
df_max_doy = df_2[df_2.month==12][['year','month','day']].groupby(['year','month'],as_index=False).max()
df_max_doy = df_max_doy.merge(df_2[['year','month','day','close']],how='inner',on=['year','month','day'])
df_min_doy = df_2[df_2.month==1][['year','month','day']].groupby(['year','month'],as_index=False).min()
df_min_doy = df_min_doy.merge(df_2[['year','month','day','close']],how='inner',on=['year','month','day'])
```

### 4.02 Create a df_yearly Data frame that has an open and close column per year.


1. The **open** column holds the closing value of the first day of January for that year.  
2. The **close** column holds the closing value of the last day of December for that year.

```python
df_min_doy = df_min_doy.rename(columns={'close':'open'})
df_min_doy.head()
df_yearly = df_min_doy[['year','open']].merge(df_max_doy[['year','close']],how='inner',on='year')
df_yearly.head()
```

### 4.03 Create df_yearly_low to hold low per year

```python
df_yearly_low = df_2[['year','low']].groupby('year',as_index=False).min()
df_yearly_low
```

### 4.04 Merge in yearly low to df_yearly

```python
df_yearly = df_yearly.merge(df_yearly_low,how='inner',on='year')
```

```python
df_yearly
```

### Create pct_low column to denote the percent change of the yearly low

```python
df_yearly['pct_low'] = df_yearly.apply(lambda r: (r.open-r.low)/r.open,axis=1)
```

### 6.00 strategies

#### 6.01 % dips

```python
upside_wait = 1000
downside_wait = -0.001
curr_pl = curr_close = 0
curr_position = 1
initial_open = float(df_2.iloc[0].open)
curr_open = initial_open
for i in range(len(df_2)):
    curr_close = float(df.iloc[i].close)
    if curr_close/curr_open - 1 >= upside_wait and curr_position==1:
        curr_pl += curr_close - curr_open
        curr_position = 0
        curr_open = curr_close
        continue
    if curr_close/curr_open - 1 <= downside_wait and curr_position==0:
        curr_position = 1
        curr_open = curr_close
        continue
if(curr_pl==0):
    curr_pl = float(df_2.iloc[-1].close) - initial_open 
print(curr_pl/initial_open)
```

### 6.03 Do a trade when there has been the same directional move for "n" number of days

```python
df_3 = df_2.copy()
# use avergage of high and low instead of close to decide if you should get in
df_3['avg_price'] = df_3.apply(lambda r: (r.high + r.low)/2,axis=1)
df_3['pct_chg'] = df_3.avg_price.pct_change()
# get whether the change was up or down that day
df_3['sign'] = df_3.pct_chg.apply(lambda p: 0 if abs(p)==0 else p/abs(p))
df_pchg = df_3[['year', 'month', 'day','open','close','pct_chg','sign']][1:]
# define the number of days that you want the continuous up's or down's to repeat before you take a position for 
#   the next day
window_size = 2
# add up the sign values, and see if, on any day, they add up to the window size
df_pchg['roll'] = df_pchg.sign.rolling(window=window_size).sum()
#  if they add up, then you will enter into a trade
df_pchg['do_trade'] = df_pchg.roll.apply(lambda x: abs(x)==window_size)
#  HOWEVER, you will enter into the trade on the NEXT day, so shift the entry indicator to the next day
df_pchg['do_trade'] = df_pchg.do_trade.shift(1)
# get the sign that tells if you bought or sold
df_pchg['prev_sign'] = df_pchg.sign.shift(1)
df_pchg['prev_close'] = df_pchg['close'].shift()
# calculate the p_l resulting from entering a position on the open, and closing it on the close
df_pchg['p_l'] = df_pchg.apply(lambda r: (r.close/r.prev_close - 1) * r.prev_sign  if r.do_trade else 0,axis=1 )
df_pchg_trade_days = df_pchg[df_pchg.p_l != 0]
pl_sum = df_pchg_trade_days.p_l.sum()
pl_std = float(df_pchg_trade_days.p_l.std())
pl_avg = float(df_pchg_trade_days.p_l.mean())
pl_sharpe = pl_avg/pl_std * 256**.5
print(f'total pl: {round(pl_sum,5)}, average pl: {round(pl_avg,5)}, stf: {round(pl_std,5)}, sharpe: {round(pl_sharpe,3)} ')
```
